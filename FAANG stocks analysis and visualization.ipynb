## FAANG Stocks Analysis and Visualization

In this project, I will analyze and visualize the FAANG companies stocks, as of in the last 5 years (from August 2016 - August 2021) 
* Facebook (NASDAQ:FB)
* Amazon (NASDAQ:AMZN)
* Apple (NASDAQ:AAPL)
* Netflix (NASDAQ:NFLX)
* Alphabet (NASDAQ:GOOG)

This project will use Pandas, Pandas-Datareader, and Matplotlib module to get data, analyze to see the change of these stocks prices in the recent 5 years. This will help investors have an idea, and might decide which company to invest for long-term.

### 1. Import Packages
Import modules needed for analyzing

```python
import pandas as pd
import numpy as np
import pandas_datareader as web
from matplotlib import pyplot as plt
```

### 2. Load the daily stock data for the FAANG stocks and inspect attributes.

#### Step 1: Define Stocks
Create a list named `symbols` containing the tickers for the FAANG stocks.
#### Step 2: Create Dates
Create a datetime object representing `start_date` and `end_date`.
#### Step 3: Retrieve Data
Call the function `web.get_data_yahoo()` to retrieve data.
#### Step 4: View Data
Inspect DataFrames to see what type of information to analyze.

```python
symbols = ['FB', 'AMZN', 'AAPL', 'NFLX', 'GOOGL']
start_date = '2016-08-01'
end_date = '2021-08-01'
stock_data = web.get_data_yahoo(symbols, start_date, end_date)
```

```python
stock_data
```

I need to take a closer look at Adjust Closing Price to analyze the changes of the stocks overtime.

```python
stock_data['Adj Close']
```

### 3. Plot the adjusted closing prices over time.

Create a plot with matplotlib that shows the adjusted closing prices of each stock over time. 

```python
# stock_data_adj_close = stock_data['Adj Close']
plt.figure(figsize=(15,8))
plt.plot(stock_data['Adj Close'], label=symbols)
plt.xlabel('Date')
plt.ylabel('Adjusted closing price')
plt.title('Adjusted closing prices over time')
plt.legend(loc='upper left')
plt.show()
```

### 4. Calculate and plot the daily simple rate of return over time.
Create a plot with matplotlib that shows the daily simple rate of return for each tech stock over time. Label the graph appropriately. 

```python
plt.figure(figsize=(15,7))
plt.plot(stock_data['Adj Close'].pct_change(), label=symbols)
plt.xlabel('Date')
plt.ylabel('Rate of Return')
plt.title('Daily Simple Rate of Return Over Time')
plt.legend(loc='upper left')
plt.show()
```

### 5. Create subplots of daily simple rate of return.
In order to better visualize the daily returns, create a subplot for each FAANG stock. 

```python
fig=plt.figure(figsize=(30,30))
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax1.plot(stock_data['Adj Close']['AMZN'].pct_change())
ax1.set_title('Amazon')
ax2.plot(stock_data['Adj Close']['AAPL'].pct_change())
ax2.set_title('Apple')
ax3.plot(stock_data['Adj Close']['FB'].pct_change())
ax3.set_title('Facebook')
ax4.plot(stock_data['Adj Close']['GOOGL'].pct_change())
ax4.set_title('Alphabet')
ax5.plot(stock_data['Adj Close']['NFLX'].pct_change())
ax5.set_title('Netflix')
plt.tight_layout()
plt.show()
```

### 6. Calculate and plot the mean of each stock's daily simple rate of return

#### Calculate mean rate of return and plot bar chart
For each stock, calculate the mean daily simple rate of return.

Based on the mean rate of return, which stock would be the best option to invest in?

```python
stock_data_daily_return = stock_data['Adj Close'].pct_change()
daily_mean = stock_data_daily_return.mean()
daily_mean
```

```python
daily_mean.keys()
```

```python
daily_mean_bar = []
for key in daily_mean.keys():
    daily_mean_bar.append(daily_mean[key])
daily_mean_bar
```

```python
plt.bar(daily_mean.keys(), daily_mean_bar)
plt.xlabel('FAANG Stocks')
plt.ylabel('Mean ROR')
plt.title('Daily Mean ROR')
plt.show()
```

Netflix has the highest mean simple rate of return over the period of data collected. Thus Netflix would have been a good choice for investment over this period of time. Google and Facebook, on the other hand, have the lowest mean simple rate of return over the period.

### 7. Calculate and plot the variance.

#### Calculate the variance and plot bar chart
For each stock, calculate the variance of the mean daily simple rate of return.

Based on the variance, which stock would be the riskiest to invest in?

```python
daily_var = stock_data_daily_return.var()
daily_var
```

```python
daily_var_bar = []
for key in daily_var.keys():
    daily_var_bar.append(daily_var[key])
daily_var_bar
```

```python
plt.bar(daily_var.keys(), daily_var_bar)
plt.xlabel('Stocks')
plt.ylabel('Variance')
plt.title('Daily Variance')
plt.show()
```

Netflix shows the highest variance of all the stocks, indicating it can be a riskier investment. Google shows the lowest variance, indicating that the returns are more predictable. This goes along with the typical understanding of higher return, high risks stocks, and lower return, low risk stocks.

### 8. Calculate and plot the standard deviation

#### Calculate the standard deviation and plot bar chart
For each stock, calculate the standard deviation of the mean daily simple rate of return.

Based on the standard deviation of the rates of return, which stock would you choose to invest in and why?

```python
daily_std = stock_data_daily_return.std()
daily_std
```

```python
daily_std_bar = []
for key in daily_std.keys():
    daily_std_bar.append(daily_std[key])
daily_std_bar
```

```python
plt.bar(daily_std.keys(), daily_std_bar)
plt.xlabel('Stocks')
plt.ylabel('Standard Deviation')
plt.title('Daily STD')
plt.show()
```

Netflix has the largest standard deviation, thus it is the most volatile stock in the FAANG companies. Volatile stocks are mostly risky stocks. Netflix also has the highest daily mean rate of return. It does support the common understanding about high risk, high rewards stocks. Google also has the lowest standard deviation, so it is the least volatile stock. Google also has the lowest mean rate of return.

### 9. Calculate the correlations

None of the stocks are negatively correlated. Microsoft and Apple have the highest correlation, while Facebook and Amazon exhibit the lowest correlation.

### 10. Calculate the return over 5 years
Calculate the return from 2016-08-01 to 2021-08-01 of each stock, and together with the variance, standard deviation, and mean rate of return to decide which stock to invest for longterm.

```python
# get the adjusted closing price of each stock on the start and end date
price_change = stock_data['Adj Close'].iloc[[0,-1]]
price_change
```

```python
# reshape the price_change DataFrame to calcuate percentage change
price_change_t = price_change.T
price_change_t.columns=['Price Start', 'Price End']
price_change_t['percentage_change'] = price_change_t['Price End']/price_change_t['Price Start']
price_change_t
```

We can see that investing in Apple will give the highest return in the time period. While investing in Facebook leaves the lowest return among the FAANG companies during the time period. 
Netflix is the riskiest, and most volatile stock, but it does not guarantee a high ROI as Apple.
Therefore, if you are a safe investor caring about ROI, you should invest in Apple stock. If you like higher risk, you can choose to invest in Netflix, for a high return.
